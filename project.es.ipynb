#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
import pandas as pd
ds=pd.read_csv('assets/real_estate.csv', sep=';')

poblacion1 = "Valdemorillo"
poblacion2 = "Galapagar"

media_precio1 = ds[ds["level5"] == poblacion1]["price"].mean()
media_precio2 = ds[ds["level5"] == poblacion2]["price"].mean()

print(f"La media de precios en {poblacion1} es {media_precio1:.2f} USD.")
print(f"La media de precios en {poblacion2} es {media_precio2:.2f} USD.")

diferencia= abs(media_precio1-media_precio2)
print (f"\nDifernecia de precios: {diferencia:.2f}")

if diferencia < 5000:
    conclusion = "Los precios promedio en Valdemorillo y Galapagar son semejantes."
elif media_precio1 < media_precio2:
    conclusion = f"La diferencia de precios es de {diferencia:.2f} USD/m² a favor de {poblacion2}, lo que indica que la media de pisos en {poblacion2} es más cara que en {población1}"
else:
    conclusion = f"La diferencia de precios es de {diferencia:.2f} USD/m² a favor de {poblacion1}, lo que indica que la media de pisos en {poblacion1} es más cara que en {población2}"

print(f"\nConclusión: {conclusion}")
```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
import pandas as pd
ds=pd.read_csv('assets/real_estate.csv', sep=';')

ds["pps"] = ds["price"] / ds["surface"]

media_valdemorillo = ds[ds["level5"] == "Valdemorillo"]["pps"].mean()
media_galapagar = ds[ds["level5"] == "Galapagar"]["pps"].mean()

print(f"Precio promedio por m² en Valdemorillo: {media_valdemorillo:.2f} USD")
print(f"Precio promedio por m² en Galapagar: {media_galapagar:.2f} USD")

diferencia=abs(media_valdemorillo - media_galapagar)
print(f"\nDiferencia: {diferencia: .2f} USD")

if diferencia < 100:  
    conclusion = "Los precios promedio por metro cuadrado en Valdemorillo y Galapagar son prácticamente iguales."
elif media_valdemorillo > media_galapagar:
    conclusion = "El precio promedio por metro cuadrado en Valdemorillo es más alto que en Galapagar."
else:
    conclusion = "El precio promedio por metro cuadrado en Galapagar es más alto que en Valdemorillo."


print(f"\nConclusión: {conclusion}")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_sur = ds[ds["level5"].isin(poblaciones_sur)].copy()

scaler = MinMaxScaler()
ds_sur["price_normalized"] = scaler.fit_transform(ds_sur[["price"]])

plt.figure(figsize=(12, 6))

colores = {
    "Fuenlabrada": "yellow",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "red"
}

for poblacion in poblaciones_sur:
    precios_norm = ds_sur[ds_sur["level5"] == poblacion]["price_normalized"]
    plt.hist(precios_norm, bins=30, alpha=0.5, color=colores[poblacion], edgecolor="black", label=poblacion)

plt.xlabel("Precio Normalizado")
plt.ylabel("Frecuencia")
plt.title("Distribución Normalizada de Precios en el Cinturón Sur de Madrid")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
from ipywidgets import HTML

import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

ds_sur = ds[ds["level5"].isin(poblaciones_sur)].dropna(subset=["latitude", "longitude"])

ds_sur["latitude"] = ds_sur["latitude"].astype(str).str.replace(',', '.').astype(float)
ds_sur["longitude"] = ds_sur["longitude"].astype(str).str.replace(',', '.').astype(float)

coordenadas = ds_sur[["level5", "latitude", "longitude"]].to_dict(orient="records")

colores = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "orange"
}

mapa = Map(center=(40.35, -3.8), zoom=11.5, basemap=basemaps.OpenStreetMap.Mapnik)

for dato in coordenadas:
    marcador = CircleMarker(
        location=(dato["latitude"], dato["longitude"]),
        radius=5,  
        color=colores[dato["level5"]],
        fill_color=colores[dato["level5"]],
        fill_opacity=0.6
    )
    marcador.popup = HTML(value=f"<b>{dato['level5']}</b><br>Lat: {dato['latitude']}<br>Lon: {dato['longitude']}")
    mapa.add_layer(marcador)

mapa



```
